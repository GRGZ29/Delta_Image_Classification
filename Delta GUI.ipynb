```python
import tkinter as tk
from efficientnet.tfkeras import EfficientNetB0, preprocess_input
from tkinter import filedialog
from tensorflow.keras.applications.imagenet_utils import decode_predictions
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
from PIL import Image, ImageTk
import numpy as np



model = tf.keras.models.load_model(r'C:/Users/RUBINOS/Desktop/Diabetic Retinopathy/Untitled12.h5')
# Create a Tkinter window
window = tk.Tk()
window.title("Demo Image Classification using EfficientNetB0 Pretrained Model")
window.geometry("500x500")
window.configure(bg='#98AFC7') 

# Load the pre-trained model
model = EfficientNetB0(weights='imagenet')

def open_file_dialog():
    file_path = filedialog.askopenfilename()
    classify_image(file_path)

def classify_image(img_path):
    # Load and preprocess the input image
    img = Image.open(img_path)
    img = img.resize((224, 224))  # Resize the image to match model input size
    img = np.array(img)
    img = np.expand_dims(img, axis=0)
    img = preprocess_input(img)

    # Make predictions
    predictions = model.predict(img)

    # Decode and print the top prediction with custom labels
    decoded_prediction = decode_predictions(predictions, top=1)[0][0]

    # Map ImageNet label to your custom category
    custom_categories = {
        'n01930112': 'Healthy',
        'n03223299': 'Mild DR',
        'n02264363': 'Moderate DR',
        'n03291819': 'Proliferative DR',
        'n03598930': 'Severe DR',
    }

    # Extract information from the decoded prediction
    imagenet_label, human_readable_label, confidence = decoded_prediction

    # Map predicted class index to custom category
    mapped_category = custom_categories.get(imagenet_label, 'Unknown')

    # Display the predictions
    result_label.config(text=f"Prediction: {human_readable_label}\nCategory: {mapped_category}\nConfidence: {confidence:.2%}")

    # Display the selected image
    img = Image.open(img_path)
    img = img.resize((300, 300), Image.ANTIALIAS)
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img

# Button to open file dialog
btn_open_file = tk.Button(window, text="Click here to Open an Image", command=open_file_dialog, bg='#001F3F', fg='white')
btn_open_file.pack(pady=10)


# Label to display the predictions
result_label = tk.Label(window, text="Prediction:")
result_label.pack(pady=10)

# Label to display the image
panel = tk.Label(window, bg='#add8e6')  # Light Blue
panel.pack()

# Run the GUI
window.mainloop()
```
